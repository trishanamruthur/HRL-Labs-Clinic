## Overview of a Single Spin and Single Quantum Dot
The Loss-DiVincenzo (LD) spin qubit is a single-electron spin qubit that consists of a direct mapping between spin operators and encoded Pauli operators. These qubits involve electrons tunneling to/from fermionic electron baths. Higher energy quantum dot spin states will allow for tunneling to the Fermi sea, and the static magnetic field of the electron bath will prevent tunneling from lower-energy spin states. The spin state of the electron, and thus the qubit state, is determined by the presence or absence of tunneling events. The Loss-DiVincenzo model serves as a basis for further developments in spin-qubits. (Based on [Burkard et al](../../../PDFs/Burkard.pdf) ). 

To further elaborate on the quantum dot itself, it is a structure filled with electrons that is coupled to electron reservoirs. As seen in both Figure 1 of Hanson and in the animation, electrons from the reservoirs from either side of dot tunnel in and out based on the Fermi energy levels. Single quantum dots involve the loading/unloading of single electrons (or coupled, discussed later) in and out from the reservoirs, a source and a drain where the flow of the electrons move from source to dot to drain, with tunnel barriers as the connection between.  


## Coulomb Blockades
The number of electrons traveling within a quantum dot is controlled via Couloomb blockade. 
"The electronic properties of quantum dots are dominated by two effects. First, the Coulomb repulsion between electrons on the dot leads to an energy cost for adding an extra electron to the dot. Due to this charging energy tunneling of electrons to or from the reservoirs can be suppressed at low temperatures." [Hanson et al, 2007](../../../PDFs/Hanson.pdf)


## Animation of Loading and Unloading Dot
This animation demonstrates the loading and unloading of a quantum dot from the electron wells, and how Fermi energy levels affect the movement (or lack thereof) of the dot. For a single quantum dot, depending on the Fermi energy levels of the source and drain and the M1 gate the electron will either move or stay in place. If the energy of the M1 gate is higher than source, electrons will not be able to move from the source, and a similar blockage occurs when the drain is at a higher Fermi energy than the gate.
