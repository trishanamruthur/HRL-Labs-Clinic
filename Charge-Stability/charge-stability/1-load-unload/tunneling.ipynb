# Electron Tunneling

## Square Potential
In Quantum Mechanics, particles can [tunnel](gls:tunneling) through classically forbidden areas where  $E \leq V.$ The [quantum dot](gls:quantum-dot) works by having potential barriers surrounding the dot. Thus, an electron tunnels through a quantum dot. The T-gates' voltage determines the height of the barrier, and therefore how hard it is for the electron to tunnel into the dot. Thus, the tunneling rate of the electron depends on the T-gates.

To understand tunneling, we can begin with a toy example. The dot is in the plunger gate which here has a constant potential $V_P$. The T-gates we simplify having the same constant voltage $V_T$. We further simplify by setting the source and drain potential of the dot having 0 voltage. Below is an equation for the potential around the dot. 

$$
V(x) = 
\begin{cases}
0 & x< 0\\
V_T & 0 < x < x_1\\
V_P & x_1 < x < x_2\\
V_T & x_2 < x < x_3\\
0 & x>x_3\\
\end{cases} 
$$


Then, we can note that the [wave function](gls:wavefxn) of the electron would satisfy [the Schrodinger equation](gls:schrodinger). To understand the procedure of solving the double-wall potential described here, [Townsend](../../../PDFs/Townsend_textbook.pdf) chapter 6.10 does a good job describing the general procedure. We will be following the procedure here. First, we can create a general wave function and solve for the coefficients using the boundary conditions. 

Thus, our wave function would be as follows. 
$$
\psi (x) = 
\begin{cases}
Ae^{ik_0x} + Be^{-ik_0x} & x<0\\
Ce^{ik_Tx}+De^{-ik_Tx} & 0<x<x_1\\
Ee^{ik_Px}+Fe^{-ik_Px} & x_1<x<x_2\\
Ge^{ik_Tx}+He^{-ik_Tx} & x_2<x<x_3\\
Ie^{ik_0x} & x>x_3
\end{cases}
$$ 

where 

$$
\begin{align*}
k_0 &= \sqrt{2mE}/\hbar\\
k_T &= \sqrt{2m(E-V_T)}/\hbar\\
k_P &= \sqrt{2m(E-V_P)}/\hbar
\end{align*}
$$

$k$ is called the wave number. It encodes the energy of the wave of the electron. We can note that $e^{ikx}$ would be a wave that travels to the right and $e^{-ikx}$ would represent a wave traveling to the left. 

Then, the tunneling probability for the electron to enter the dot would be then 

$$
T_{\text{enter}} = \frac{|E|^2}{|A|^2}
$$

this is because $|A|$ encodes the probability that the electron is traveling to the right in the region before the dot and $|E|$ encodes the probability that the electron crosses the potential barrier and enters the dot. 

Similarly, the tunneling probability for the electron to enter the dot would be 

$$
T_{\text{exit}} = \frac{|I|^2}{|E|^2}
$$

Thus, we can see that to find the tunneling probability, we need to find the coefficients of the wave function. To do this, we use the fact that the wave function and its derivative must be continuous and the boundaries to get the following system of equations. 

Thus, at $x=0$, we get
$$
A+B=C+D\\
ik_0(A-B)=ik_T(C-D)
$$

Similarly, we would get more equations at each of the boundaries. Then, we need to solve the system of equations to get the tunneling probability. We can solve this numerically. 
Here we can see a slider that calculates the tunneling probabilities for different voltages of the T-gates.

 **ADD SLIDER HERE**
 
Thus, we can see that the T gates directly impact the probability that an electron enters and exits a quantum dot. 

 

## Non-Square Potential & WKB Approximation
We can note that the previous section was a rather simplistic view of the potentials. We assumed that the potential was constant in regions and that the barriers were perfect squares. However, the potential of the gates is not a constant line and thus we can use a different method of finding the tunneling probability.

From [Davies et al 1995](https://doi.org/10.1063/1.359446), the gate potential, assuming a finite rectangle gate, is given by 

$$ 
\phi_G(r,d) = V\cdot (g(x-L, y-B)+g(x-L, T-y)+g(R-x, y-B)+g(R-x, T-y))
$$ 

where the dimensions of the gate are $L<x<R$ and $B<y<T$ and $V$ is the voltage at the boundaries of the gate and where

$$
g(u,v) = \frac{1}{2\pi} \text{arctan} \left(\frac{uv}{dR}\right); R= \sqrt{u^2+v^2+d^2}
$$

Then, we can combine these gate potential functions to create the potential, V(x), of the triple gate system described earlier. 
Since the potential is no longer square, we can use the process of WKB approximation to solve the tunneling probability where now 

$$
T = \text{exp}\left(-2 \int dx \sqrt{\frac{2m}{\hbar^2}[V(x)-E]}\right)
$$

We would obtain this tunneling probability numerically. 

**CREATE SLIDER FOR THIS MAYBE?**

### Better Approximations, the Fermi-Hubbard Model
The WKB approximation is a simple approximation of the tunneling probability. However, the approximation relies on the fact that the potential does not change over time. However, these potentials might change as the voltage of the gates change. Thus, a more sophisticated model is necessary to model the tunneling probability. This is the Fermi-Hubbard Model which deals with particles in 2D or 3D lattices . To find out more about the model, [Esslinger 2010](https://arxiv.org/pdf/1007.0012.pdf#:~:text=Figure%202%3A%20A%20Hubbard%20model,atom%20of%20a%20different%20spin.) does a good job explaining the basics. 

**TODO expand on the Fermi-Hubbard model?**
